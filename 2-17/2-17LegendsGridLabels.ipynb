<h1 align="center" style="color:purple">Matplotlib Tutorial: Labels, Legend, Grid</h1>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[32,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

<h2 style="color:purple">Labels</h2>

```python
# set axes labels and chart title
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend()
```

<h2 style="color:purple">Legend</h2>

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
# Show legend
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='best',shadow=True)
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
# Legend at different location with shadow enabled and fontsize set to large
plt.plot(days, max_t, label="max")
plt.plot(days, avg_t, label="average")
plt.plot(days, min_t, label="min")

plt.legend(loc='upper right',shadow=True,fontsize='large')
```

<h2 style="color:purple">Grid</h2>

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label="max",color="red")
plt.plot(days, avg_t, label="average")
plt.plot(days, min_t, label="min")

plt.legend(loc='lower left', fontsize="large",shadow=True)
plt.grid()
```
