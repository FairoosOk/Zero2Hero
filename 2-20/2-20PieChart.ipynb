<h1 align="center", style="color:purple">Matplotlib Tutorial: Pie Charts</h1>

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

<h3 style="color:purple">Draw pie chart to track down home expenses</h3>

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

<h3 style="color:purple">Draw pie chart as perfect circle</h3>

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

<h3 style="color:purple">Show percentages for every pie. Specify radius to increase chart size</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.2f%%',radius=2)
plt.show()
```

<h3 style="color:purple">Explode</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.3,0.2])
plt.show()
```


<h3 style="color:purple">counterclock and angle properties</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,counterclock=False,explode=[0,0,0,0.1,0.2], startangle=90)
plt.show()
```

More Ref: https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.pie.html

<h2>You can also save in pdf format and to a different location on your computer by specifying absolute path<h2>

```python
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.2,0.1,0],shadow=True)
plt.savefig("c:/github/expensechart.pdf", bbox_inches="tight",pad_inches=4, transparent=True)
```
