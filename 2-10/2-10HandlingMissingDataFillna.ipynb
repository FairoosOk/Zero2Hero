# <font color="purple"><h4 align="center">Handling Missing Data - fillna</font>

```python
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
df
```

```python
df.set_index('day',inplace=True)
df
```

# <font color="purple">fillna</font>

<font color="purple">**Fill all NaN with one specific value**</font>

```python
new_df = df.fillna(0)
new_df
```

<font color="purple">**Fillna using Column names and Dictionary**</font>

```python
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

<font color="purple">**Use method to determine how to fill na values**</font>

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method="bfill")
new_df
```

<font color="purple">**Use of axis**</font>

```python
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df
```

<font color="purple">**limit parameter**</font>

```python
new_df = df.fillna(method="ffill",limit=2)
new_df
```
