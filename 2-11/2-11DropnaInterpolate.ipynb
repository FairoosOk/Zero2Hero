# <font color="purple"><h4 align="center">Handling Missing Data Interpolate, Dropna</font>

```python
import pandas as pd
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
df
```

```python
df.set_index('day',inplace=True)
df
```

### <font color="red">Interpolate</font>

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method="time") 
new_df
```

**Notice that in above temperature on 2017-01-04 is 29 instead of 30 (in plain linear interpolate)**

**There are many other methods for interpolation such as quadratic, piecewise_polynomial, cubic etc. 
Just google "dataframe interpolate" to see complete documentation**

### <font color="red">dropna()</font>

```python
df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

###### at least 1 non na values-- then don't remove raw

```python
new_df = df.dropna(thresh=2)
new_df
```

### <font color="red">Inserting Missing Dates</font>

```python
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df.reindex(idx)
```
