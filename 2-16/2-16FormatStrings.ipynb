<h1 align="center" style="color:purple">Matplotlib Tutorial: pyplot.plot format strings and other options</h1>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,color="red",marker="*",linestyle="dashed")
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'m>')
```

```python
# Same effect as 'b+' format string
plt.plot(x,y,color='#00EE00',marker='+',linestyle='')
```

<h3 style="color:purple">Marker Size</h3>

```python
plt.plot(x,y,'b+-.',markersize=10)
```

<h3 style="color:purple">alpha property to control transparency of line chart</h3>

```python
plt.plot(x,y,'g<',alpha=.5) # alpha can be specified on a scale 0 to 1
```
