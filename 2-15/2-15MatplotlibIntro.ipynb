```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6]
y=[50,51,52,48,47,46]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color="indigo",linewidth=3,linestyle=":")
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Whether')
plt.plot(x,y,color='indigo',linestyle="dotted")
```
